Creamos un archivo "Customers" y lo rellenamos con la biblioteca Faker para llenar la tabla de datos aleatorios.

```python
import pandas as pd
from faker import Faker

fake = Faker()
num_filas = 1000

columnas = ['Customer_ID', 'First_Name', 'Last_Name', 'Phone_Number', 'Email', 'Address']
data = []
customer_id_sequence = itertools.count(start=1, step=1)

for _ in range(num_filas):
    customer_id = next(customer_id_sequence)
    data.append({
        'Customer_ID': customer_id,
        'First_Name': fake.first_name(),
        'Last_Name': fake.last_name(),
        'Phone_Number': fake.phone_number(),
        'Email': fake.email(),
        'Address': fake.address()
    })

df = pd.DataFrame(data, columns=columnas)

df.to_csv('customers.csv', index= False)
```
